### Tipos de dados, conversão e padronização
O trabalho acima feito foi a conversão do DataFrame 'user_score' de object para float transformando todos os números de avalições dos usuários em valores numéricos reais, e não texto como estava. Essa conversão é essencial para análise quantitativa. O tipo float permite o cálculo de métricas estatísticas (média, mediana e correlação) e a criação de visualizações como histogramas e boxplots, que seriam impossíveis com o tipo object (tipo texto).

Outra conversão realizada foi o DataFrame 'year_of_release' que estava em float e foi convertido para Int. Como ele representa o ano de lançamento do jogo faz sentido que seja um número inteiro. Essa mofificação foi feita para garantir que o ano de lançamento seja tratado como valor numérico, permitindo operações temporais e filtragens. Vale ressaltar que o tipo Int64 (com "I" maiúsculo) foi utilizado porque ele permite armazenar valores ausentes (NaN), diferentemente do tipo int64 tradicional. Assim, se mantém os NaNs preservando a informação de falta de dados.



<div class="alert alert-block alert-info">
<b>Resposta do Aluno: no caso da coluna 'year_of_release', optei por imputar o ano de lançamento do mesmo jogo em outra plataforma em vez de manter valores ausentes (NaN), pois o mesmo jogo (name) aparece em outra plataforma com ano definido. Essa abordagem é adequada porque o ano de lançamento costuma ser o mesmo (ou muito próximo) entre plataformas, o que torna a imputação coerente e reduz perdas de informação nas análises temporais (ano de lançamento). O processo é realizado de forma automatizada e controlada, com registro dos casos imputados. </b> <a class="tocSkip"></a>

```python
# Criar série com ano por name (ignora NaN)
year_mode_by_name = (df.dropna(subset=['year_of_release'])
                       .groupby('name')['year_of_release']
                       .agg(lambda s: pd.Series.mode(s)[0] if not s.mode().empty else s.median()))

# Função para imputar
def impute_year(row):
    if pd.isna(row['year_of_release']):
        v = year_mode_by_name.get(row['name'], pd.NA)
        return int(v) if pd.notna(v) else pd.NA
    return row['year_of_release']

# Aplicar (preservar tipo Int64)
df['year_of_release_before'] = df['year_of_release']
df['year_of_release'] = df['year_of_release'].astype('Int64')
df['year_of_release'] = df.apply(impute_year, axis=1).astype('Int64')

# Flag para identificar imputados (True se antes era NaN e agora tem valor)
df['year_imputed'] = df['year_of_release_before'].isna() & df['year_of_release'].notna()

# Ver resumo
df['year_imputed'].value_counts(dropna=False)
```



<div class="alert alert-block alert-info">
<b>Resposta do Aluno: quanto as colunas 'score' e 'rating' é melhor deixar NaNs nesse caso. Para as colunas relacionadas a avaliações ('user_score', 'critic_score') e classificação etária ('rating'), a decisão é de manter os valores ausentes (NaN) sem imputação. Isso ocorre porque esses campos representam avaliações subjetivas e classificações oficiais, que não podem ser inferidas de forma confiável a partir de outros registros.
A imputação artificial nesses casos comprometeria a integridade das análises estatísticas e dos testes de hipótese. Portanto, a melhor decisão é por preservar os NaNs e tratar os dados faltantes apenas na fase analítica.  </b> <a class="tocSkip"></a>
</div>




```python

# Garantir 'user_score' numérico (coerce já feito antes, repetir para segurança)
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')

# Garantir 'critic_score' numérico
df['critic_score'] = pd.to_numeric(df['critic_score'], errors='coerce')

# Mostrar quais foram NaN antes 
df['critic_missing'] = df['critic_score'].isna()
df['user_missing'] = df['user_score'].isna()

# Resumo
df[['critic_missing','user_missing']].agg(['sum','mean']).T
```

<b>Resposta do Aluno: para as colunas categóricas 'genre' (gênero) e 'name' (nome do jogo) por serem strings a melhor abordagem é preencher os valores ausentes com a categoria 'Desconhecido'. Isso evita a perda de registros potencialmente relevantes e mantém a integridade das análises agregadas (como contagens, agrupamentos e comparações por gênero ou título). Diferentemente de variáveis numéricas ou avaliativas, a imputação aqui não distorce o significado estatístico, pois 'Desconhecido' funciona como uma categoria neutra, permitindo contabilizar esses jogos sem atribuir um valor arbitrário. Assim, a base permanece completa e coerente para as etapas de análise descritiva e visualização.  </b> <a class="tocSkip"></a>

```python
# Verificar valores ausentes antes do preenchimento
print("Valores ausentes antes do preenchimento:")
print(df[['genre', 'name']].isna().sum())

# Preencher valores ausentes com 'Desconhecido'
df['genre'] = df['genre'].fillna('Desconhecido')
df['name'] = df['name'].fillna('Desconhecido')

# Verificar novamente após o preenchimento
print("\nValores ausentes após o preenchimento:")
print(df[['genre', 'name']].isna().sum())

# Registrar quantos valores foram preenchidos
filled_genre = df['genre'].value_counts().get('Desconhecido', 0)
filled_name = df['name'].value_counts().get('Desconhecido', 0)

print(f"\nRegistros com 'Desconhecido' em genre: {filled_genre}")
print(f"Registros com 'Desconhecido' em name: {filled_name}")
```

## Análise exploratória de dados - Desvendando o Mercado: Tendências de Vendas, Longevidade de Plataformas e Correlação de Avaliações.

A análise exploratória dos dados é dedicada a identificar padrões de comportamento do mercado e selecionar o período mais relevante para a modelagem preditiva. Inicialmente, analiso o lançamento anual de jogos para estabelecer a significância dos dados históricos. Em seguida, faço um exame detalhado da variação das vendas por plataforma, determinando o ciclo de vida das tecnologias de jogos (ascensão e declínio) e definindo o período de análise ideal para construir um modelo para o ano seguinte.

As plataformas mais vendidas são isoladas para estudo, comparando seu crescimento e declínio a fim de selecionar as platafromas mais lucrativas. A análise de vendas é aprofundada com a construção de um diagrama de caixa para comparar a distribuição global de vendas entre plataformas, seguida pela investigação da correlação entre as avaliações de usuários e críticos e o desempenho de vendas em uma plataforma de sucesso. 

Por fim, a distribuição de jogos por gênero é examinada para identificar os tipos de jogos mais e menos lucrativos, culminando em insights críticos sobre quais características impulsionam o sucesso no mercado global de videogames.

```python
# Contar quantos jogos foram lançados em cada ano
games_per_year = df['year_of_release'].value_counts().sort_index()
display(games_per_year)
```

```python
# Visualizar em gráfico de linha para entender a tendência de jogos lançados por ano
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
games_per_year.plot(kind='bar')
plt.title('Número de jogos lançados por ano')
plt.xlabel('Ano de lançamento')
plt.ylabel('Quantidade de jogos')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

O gráfico mostra a evolução da quantidade de jogos lançados entre 1980 e 2016.
Na minha análise, podemos dividir o comportamento em quatro grandes períodos no mercado.

1- **Início do mercado (1980–1994)** — Dados escassos e mercado em formação.
Nessa fase, o número de jogos lançados por ano é muito baixo, com valores que raramente ultrapassam algumas dezenas. Isso reflete o estágio inicial da indústria. Além disso, os dados históricos são incompletos, o que explica parte da baixa contagem. **Esses anos não são estatisticamente significativos para análises comparativas, pois a coleta digital de dados ainda estava no começo.**

2- **Expansão e consolidação (1995–2005)** — Crescimento acelerado.
A partir de meados dos anos 1990, há um crescimento constante no número de lançamentos. Esse movimento coincide com a popularização de videogames. **A globalização da indústria de jogos passa a ter lançamentos simultâneos no mundo. Esse período marca o início da era moderna dos videogames e uma expansão significativa da base de consumidores.**

3- **Auge da produção (2006–2009)** — Ponto máximo de lançamentos. **O número de títulos atinge seu pico em 2008 e 2009, com mais de 1.400 jogos lançados por ano.** Esse auge está fortemente associado à geração de videogames que revolucionaram a experiência de jogo e atraíram muitas desenvolvedoras. **Esse é o período mais representativo da base de dados, refletindo o auge da produção comercial global.**

4- **Queda gradual (2010–2016)** — Consolidação digital e mudança de modelo. **Após 2009, observa-se uma queda significativa no número de lançamentos anuais. As principais razões prováveis são a transição para o modelo digital: muitos jogos passaram a ser lançados em plataformas online e o crescimento dos jogos mobile e online, que não fazem parte do mesmo mercado de videogames.** O que ocorre com a queda gradual também tem ligação direta na mudança no foco das desenvolvedoras, priorizando títulos maiores e mais complexos (menos jogos, mas com maior investimento e tempo de desenvolvimento).

**Conclusão analítica** - O número de lançamentos de jogos cresceu de forma consistente até atingir o auge em 2008–2009, período de consolidação das platafromas dessa geração e explosão da indústria global. A partir de 2010, há uma queda gradual, influenciada por transformações no modelo de distribuição bem como no comportamento dos consumidores e na estrutura da indústria. Portanto, **para análises mais representativas, o período de 1995 a 2016 é o mais confiável — abrangendo tanto o crescimento quanto o declínio da produção tradicional de jogos.**

```python
# Calcular as vendas globais totais por plataforma
top_platforms = df.groupby('platform')['total_sales'].sum().sort_values(ascending=False).head(10)
top_platforms
```

```python
# Selecionar apenas as plataformas para visualização
df_top = df[df['platform'].isin(top_platforms.index)]
```

```python
# Agrupar as vendas por ano de lançamento e plataforma, somando as vendas globais
sales_by_year_platform = (
    df_top.groupby(['year_of_release', 'platform'])['total_sales']
    .sum()
    .reset_index()
)
```

```python
# Criar gráfico de linhas para variação das vendas (1995-2016) de cada plataforma
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
for platform in sales_by_year_platform['platform'].unique():
    data = sales_by_year_platform[sales_by_year_platform['platform'] == platform]
    plt.plot(data['year_of_release'], data['total_sales'], label=platform)

plt.title('Vendas Globais por Plataforma (1995–2016)')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Vendas Globais (em milhões)')
plt.legend()
plt.show()
```

# Etapa 5 
## Teste de Hipóteses: Comparação de Ratings de Usuários entre Plataformas e Gêneros.

Nessa etapa o trabalho é o teste de hipóteses usando as classificações médias dos usuários tanto para plataformas (Xbox One e PC) quanto para gêneros (Action e Sports). O caso clássico do teste de hipótese para diferença entre duas médias independentes tem como método ideal o teste t de Student para amostras independentes.


### Hipótese - classificações médias dos usuários das plataformas Xbox e PC.
Aplico o teste t de Student para amostras independentes (com correção de Welch), com o objetivo de verificar se existe diferença significativa entre as médias das avaliações dos usuários nas plataformas Xbox One e PC.

```python
import pandas as pd
from scipy import stats

# Filtrar apenas as plataformas desejadas
xbox_scores = df[df['platform'] == 'XOne']['user_score'].dropna()
pc_scores = df[df['platform'] == 'PC']['user_score'].dropna()

# Teste de normalidade (para verificar se o teste t é adequado)
stat_xbox, p_xbox = stats.shapiro(xbox_scores)
stat_pc, p_pc = stats.shapiro(pc_scores)

print(f"Teste de normalidade Xbox One: p-valor = {p_xbox:.4f}")
print(f"Teste de normalidade PC: p-valor = {p_pc:.4f}")
```

```python
# Teste t de Student para amostras independentes
t_stat, p_value = stats.ttest_ind(xbox_scores, pc_scores, equal_var=False)  # equal_var=False usa Welch correction

print(f"Estatística t = {t_stat:.4f}")
print(f"p-valor = {p_value:.4f}")

# Conclusão
alpha = 0.05
if p_value < alpha:
    print("Rejeitamos H₀: As médias são significativamente diferentes.")
else:
    print("Não rejeitamos H₀: Não há diferença significativa entre as médias.")
```

<div class="alert alert-block alert-danger">
<b> Comentário do revisor: </b> <a class="tocSkip"></a>

Precisamos seguir estes passos para o teste de hipótese:
- Definir a hipótese nula (ex: a altura média é igual entre homem e mulheres)
- Definir a hipótese alternativa (ex: a altura média não é igual entre homem e mulheres)
- Avaliar se as variâncias são iguais, por exemplo, calculando a variância da amostra e verificando a igualdade (ex: sample1.var() == sample2.var())
- Em seguida, ajustar equal_var=True no teste T se as variâncias forem iguais, senão usamos equal_var=False no teste T: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html.
- Avaliar o resultado do teste T

</div>


<div class="alert alert-block alert-info">
<b>Resposta do Aluno: meu objetivo é comparar se as médias das classificações dos usuários ('user_score') são iguais ou diferentes entre duas amostras: plataformas: Xbox One vs PC; e gêneros: Action vs Sports. Para isso defino as hipóteses para as plataformas Xbox Onde e PC, antes de aplicar o teste t estatístico:                              
    

Hipótese nula (H₀): As médias das classificações dos usuários são iguais entre os dois grupos.

Hipótese alternativa (H₁): As médias das classificações dos usuários são diferentes entre os dois grupos. <a class="tocSkip"></a>
</div>



```python
import scipy.stats as stats

# Filtrar os dados
xbox_one_scores = df.loc[df['platform'] == 'XOne', 'user_score'].dropna()
pc_scores = df.loc[df['platform'] == 'PC', 'user_score'].dropna()

# Converter para numérico (caso ainda tenha string)
xbox_one_scores = pd.to_numeric(xbox_one_scores, errors='coerce')
pc_scores = pd.to_numeric(pc_scores, errors='coerce')

# Remover NaN após conversão
xbox_one_scores = xbox_one_scores.dropna()
pc_scores = pc_scores.dropna()

# Calcular variâncias
var_xbox = xbox_one_scores.var()
var_pc = pc_scores.var()

print(f"Variância Xbox One: {var_xbox:.4f}")
print(f"Variância PC: {var_pc:.4f}")

# Verificar se as variâncias são semelhantes (razão)
ratio = var_xbox / var_pc if var_pc != 0 else 0
print(f"Razão entre variâncias: {ratio:.2f}")
```

```python
# Se a razão estiver próxima de 1, variâncias são semelhantes → equal_var=True
equal_var = 0.8 <= ratio <= 1.2  # intervalo de tolerância de 20%

# Aplicar o teste t
t_stat, p_value = stats.ttest_ind(xbox_one_scores, pc_scores, equal_var=equal_var)

print(f"\nTeste T:")
print(f"Estatística t = {t_stat:.4f}")
print(f"p-valor = {p_value:.4f}")
print(f"Variâncias consideradas {'iguais' if equal_var else 'diferentes'} no teste.")
```
